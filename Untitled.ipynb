```python
os.makedirs('data_png', exist_ok=True)
```

```python
import matplotlib.pyplot as plt
```

```python
_ixes = train_0.image_id
for el in range(train_0.shape[0]):
    _name = _ixes[el]    
    _img = train_0.loc[el, :].drop('image_id').values.reshape(137, 236)
    _img = _img.astype(np.uint8)
    cv2.resize(_img, (128, 128))
    cv2.imwrite(os.path.join('data_png', '{}.png'.format(_name)) , _img)
```
